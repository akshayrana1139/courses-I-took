```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
tips = sns.load_dataset('tips')
tips.head() 
fig, axes = plt.subplots(2,2, figsize=(12, 10)) 
sns.stripplot(x="day", y="total_bill", data=tips, ax=axes[0][0]) 
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True, ax=axes[0][1])
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1',split=True, ax=axes[1][0]) 
```
